```julia
using Plots, BenchmarkTools, LinearAlgebra, LaTeXStrings
gr()
```

```julia
ff(t) = t^(0.94)
ff2(t) = t^(1.06)
x1 = 0.0:0.01:7.0
x2 = 0.0:0.01:6.0

a1=6^(1.0/0.94)

p1=plot(x1,ff.(x1), aspect_ratio=:equal, frame=:origin, label=nothing, xticks=([6.0, ], [L"a", ]) ,
    yticks=([6.0, ], [L"b", ]),xlim= (-0.3, 7.5), ylim = (-0.3, 7.5),  tickfontsize = 12,
    lw=2, lc=:red)
plot!(x2, ff.(x2), lw=0, fill = (0, 0.2, :red), label = nothing)
x3 = 0.0:0.01:a1
plot!(x3, ff.(x3), lw=0, fill = (6.0, 0.2, :yellow), label = nothing)

plot!([0, 7.5], [0.0, 0.0], arrow=(:closed, 2.0), lw=1, lc=:black, label=:none)
plot!( [0.0, 0.0], [0, 7.5],arrow=(:closed, 2.0), lw=1, lc=:black, label=:none)

annotate!(7.5, -0.2, text(L"x", :black, :hcenter, :vcenter, 11))
annotate!(-0.2, 7.5, text(L"y", :black, :hcenter, :vcenter, 11))

annotate!(4, 1.5, text(L"$\int_0^a x^{p-1}\, dx$", :black, :hcenter, :vcenter, 12))
annotate!(2, 4, text(L" \int_0^b y^{q-1}\, dy", :black, :hcenter, :vcenter, 12))
annotate!(-0.3, -0.3, text(L"O", :black, :hcenter, :vcenter, 12))

a2=6^(1.0/1.06)

p2=plot(x1,ff2.(x1), aspect_ratio=:equal, frame=:origin, label=nothing, xticks=([6.0, ], [L"a", ]) ,
    yticks=([6.0, ], [L"b", ]),xlim= (-0.3, 7.5), ylim = (-0.3, 7.5),  tickfontsize = 12,
    lw=2, lc=:red)
plot!(x2, ff2.(x2), lw=0, fill = (0, 0.2, :red), label = nothing)
x3 = 0.0:0.01:a2
plot!(x3, ff2.(x3), lw=0, fill = (6.0, 0.2, :yellow), label = nothing)
plot!([0, 7.5], [0.0, 0.0], arrow=(:closed, 2.0), lw=1, lc=:black, label=:none)
plot!( [0.0, 0.0], [0, 7.5],arrow=(:closed, 2.0), lw=1, lc=:black, label=:none)

annotate!(7.5, -0.2, text(L"x", :black, :hcenter, :vcenter, 11))
annotate!(-0.2, 7.5, text(L"y", :black, :hcenter, :vcenter, 11))

annotate!(4, 1.5, text(L"$\int_0^a x^{p-1}\, dx$", :black, :hcenter, :vcenter, 12))
annotate!(2, 4, text(L" \int_0^b y^{q-1}\, dy", :black, :hcenter, :vcenter, 12))
annotate!(-0.3, -0.3, text(L"O", :black, :hcenter, :vcenter, 12))

plot(p1, p2, dpi=300)
# savefig("young_inequality.png")
```
