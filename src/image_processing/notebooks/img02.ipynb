```julia
x = -:0.01:
T = 2*
y = collect(x)
y[1]=y[end]=0.0


a1 = (2/(2*) * (y .* sin.(x)) |> sum) * 0.01
f1 = a1 .* sin.(x)
a2 = (2/(2*) * (y .* sin.(2 .* x)) |> sum) * 0.01
f2 = f1 .+ a2 .* sin.(2*x)
a3 = (2/(2*) * (y .* sin.(3 .* x)) |> sum) * 0.01
f3 = f2 .+ a3 .* sin.(3*x)
a4 = (2/(2*) * (y .* sin.(4 .* x)) |> sum) * 0.01
f4 = f3 .+ a4 .* sin.(4*x)
p1 = plot(x, y, label = L"f(x)", aspect_ratio = :equal, lw=2)
plot!(x, f1, label = L"n = 1", lw=1.5, )
plot!(x, f2, label = L"n = 2", lw=1.5, )
plot!(x, f3, label = L"n = 3", lw=1.5, )
plot!(x, f4, label = L"n = 4", lw=1.5, ls = :dash,)

x2 = -:0.01:
T = 2*
y2 =  .- abs.(x2)
a0 = (1/(2*) .* y2 |> sum) *0.01
a1 = (2/(2*) * (y2 .* cos.(x2)) |> sum) * 0.01
b1 = (2/(2*) * (y2 .* sin.(x2)) |> sum) * 0.01
f1 = a0 .+ a1 .* cos.(x2) .+ b1 .* sin.(x2)
a2 = (2/(2*) * (y2 .* cos.(2 .* x2)) |> sum) * 0.01
b2 = (2/(2*) * (y2 .* sin.(2 .* x2)) |> sum) * 0.01
f2 = f1 .+ a2 .* cos.(2 .* x2) .+ b2 .* sin.(2 .* x2)
a3 = (2/(2*) * (y2 .* cos.(3 .* x2)) |> sum) * 0.01
f3 = f2 .+ a3 .* cos.(3*x2)
a4 = (2/(2*) * (y2 .* cos.(4 .* x2)) |> sum) * 0.01
f4 = f3 .+ a4 .* cos.(4*x2)
p2 = plot(x2, y2, label = L"f(x)", aspect_ratio = :equal, lw=2)
plot!(x2, f1, label = L"n = 1", lw=1.5, )
plot!(x2, f2, label = L"n = 2", lw=1.5, )
plot!(x2, f3, label = L"n = 3", lw=1.5, )
plot!(x2, f4, label = L"n = 4", lw=1.5, ls = :dash,)

plot(p1, p2, dpi=300)
#savefig("fourier_expansion.png")
```

```julia
function _dft(f::Vector{T}, inverse = false) where T<:Union{AbstractFloat, Complex}
    N = length(f)
    if T <:AbstractFloat
        F = zeros(Complex{T}, N)
    else 
        F = zeros(eltype(f), N)
    end
    if inverse
        for i in 1:N
            F[i] = [f[k] * exp(2.0im * * (i-1) * (k-1)/N) for k  1:N]./N |> sum
        end
    else     
        for i in 1:N
            F[i] = [f[k] * exp(-2.0im * * (i-1) * (k-1)/N) for k  1:N] |> sum
        end
    end

    return F
end

dft = v-> _dft(v, false)
idft = v-> _dft(v, true)
```
